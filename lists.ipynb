## Array vs. Lists

In many programming languages, what are known as lists are often referred to as arrays. However, there are some key differences between arrays and lists.

- **Fixed vs. Dynamic Size**: Arrays have a fixed size, meaning you must specify the number of items you will store in the array before using it. In contrast, lists do not require you to specify the size in advance. You can add items to a list dynamically, with no limit. For example, when using an array, you might write `int arr[50]`, which specifies that the array will hold 50 items. If you end up using only 40 cells, the remaining 10 blocks of memory are wasted. This is not the case with lists, making them more flexible.

- **Convenience -> Heterogeneous**: Arrays are homogeneous, meaning they can only store items of the same data type. Lists, on the other hand, are heterogeneous, allowing you to store different data types within the same list.

### Disadvantages

- **Speed of Execution**: Programs using arrays generally execute more quickly compared to those using lists.

- **Memory**: Lists typically occupy more space in memory compared to arrays.

### How Do Dynamic Arrays Work?

- We start with a fixed-size array and add items to it.
- As soon as the array is full, a new array is created, double the size of the previous array. The items from the previous array are then copied into this new array. This process is repeated whenever the array fills up again.




### How Lists are Stored in Memory

When you create a list in Python, the list itself is an object stored in memory, and each element in the list is a reference (or pointer) to another object in memory. Let's break down how the list `L = [1, 2, 3]` is stored:

1. **List Object**:
   - The list `L` is an object stored at a specific memory address. This is the memory location of the list itself.
   - When you use `id(L)`, you get the memory address of the list object.

2. **Elements in the List**:
   - Each element in the list `L` is not stored directly inside the list object. Instead, the list contains references (memory addresses) to where these elements are stored.
   - When you use `id(L[0])`, `id(L[1])`, or `id(L[2])`, you get the memory addresses of the objects `1`, `2`, and `3`, respectively.

3. **Memory Sharing**:
   - Notice that the memory addresses for `L[0]` and `1`, `L[1]` and `2`, and `L[2]` and `3` are the same. This indicates that the list elements are pointing to the same objects `1`, `2`, and `3` that are stored elsewhere in memory.

### Why This Affects Speed and Drawbacks

#### Indirect Access
Since the list only stores references to the actual data, accessing an element requires an additional step of following the reference to find the object in memory. This indirect access can add a slight overhead, especially for large lists or when accessing elements repeatedly in loops.

#### Cache Efficiency
Modern CPUs use caches to speed up data access. If the elements were stored directly in the list, they could potentially be located closer together in memory, leading to better cache performance. However, since the list stores references, the actual data may be scattered across memory, leading to poorer cache efficiency.

#### Memory Fragmentation
The separation of the list and its elements can lead to memory fragmentation, where memory is used less efficiently. This can result in increased memory usage and slower performance over time, particularly in long-running programs.

These factors can make lists less optimal in scenarios where speed and memory efficiency are critical, such as real-time applications or systems with limited resources.

### Shallow Copy vs. Deep Copy

#### Shallow Copy
A **shallow copy** creates a new list, but the elements in the new list are references to the objects in the original list. This means that changes to mutable elements (like lists within lists) in the copied list will affect the original list because they both refer to the same objects.

#### Example of Shallow Copy

```python
# Original list
L = [2, 1, 5, 7, 0]
print(L)          # Output: [2, 1, 5, 7, 0]
print(id(L))      # Memory address of list L

# Shallow copy of L
L1 = L.copy()
print(L1)         # Output: [2, 1, 5, 7, 0]
print(id(L1))     # Memory address of list L1


### Deep Copy

A **deep copy** creates a new list and recursively copies all objects found within the original list. This means that both the list structure and its contents are completely independent of the original list. Changes to the copied list do not affect the original list, even if the list contains nested lists or other mutable objects.

#### Example of Deep Copy




````python
import copy
# Original list with a nested list
L = [2, 1, [5, 7], 0]
print(L)          # Output: [2, 1, [5, 7], 0]
print(id(L))      # Memory address of list L
print(id(L[2]))   # Memory address of the nested list [5, 7]

# Deep copy of L
L1 = copy.deepcopy(L)
print(L1)         # Output: [2, 1, [5, 7], 0]
print(id(L1))     # Memory address of list L1
print(id(L1[2]))  # Memory address of the nested list in L1
````

List Comprehension provides a concise way of creating lists.

Advantages of List Comprehension

#### add 1 to 10 numbers to a list

````python
 # this is what we would do
    
L = []

for i in range(1,11):
    L.append(i)

print(L)
````

````python
# same thing we can do in one line

L = [i for i in range(1,11)]
print(L)
````

````python
v = [2,3,4]
s = -3

x = []
for i in v:
    x.append(i*s)
print(x)
````

````python
# using list comprehension
v = [2,3,4]
s = -3

[s*i for i in v]
````

### Disadvantages of Python Lists

- **Slow**: Python lists can be slower compared to other data structures, especially when dealing with large datasets.
- **Risky Usage**: Python lists are mutable, which means their content can be changed after creation. This mutability can lead to unintended side effects, making them riskier to use in certain contexts.
- **Memory Consumption**: Python lists consume more memory compared to arrays and other optimized data structures.


````python
a = [1,2,3]
b = a

print(a)
print(b)

a.append(4)
print(a)
print(b)

# lists are mutable
````

- When we write `a = b`, both `a` and `b` point to the same memory location, referring to the same list.
- If we make changes to the list through `a`, the same changes will be reflected in `b`.
- To avoid this, we can use the `copy()` method. This way, even if we make changes to `a`, it won't affect `b`.


````python
a = [1,2,3]
b = a.copy()

print(a)
print(b)

a.append(4)
print(a)
print(b)

# lists are mutable
````
